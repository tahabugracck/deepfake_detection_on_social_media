## Model Geliştirme Süreci

Proje kapsamında, derin öğrenme temelli bir görüntü sınıflandırma modeli geliştirilerek deepfake tespiti gerçekleştirilmiştir. Geliştirilen model, sahte ve gerçek görseller arasında ayrım yapabilen ikili (binary) sınıflandırıcı olarak tasarlanmıştır. Bu süreçte, eğitim verisi olarak farklı kaynaklardan toplanmış gerçek ve sahte görüntüler kullanılmış, bu görseller ön işleme adımlarından geçirilerek modele uygun hale getirilmiştir. Görseller, sabit boyuta ölçeklenmiş, normalize edilmiş ve sınıf etiketleri ile birlikte modele beslenmiştir.

Model mimarisi, evrişimli sinir ağları (CNN) kullanılarak oluşturulmuştur. Katmanlar arasında filtre boyutları, aktivasyon fonksiyonları ve havuzlama işlemleriyle modelin derinliği ve genelleme kabiliyeti optimize edilmiştir. Eğitim aşamasında, overfitting’i önlemek adına çeşitli düzenlileştirme stratejileri denenmiş, hiperparametreler kademeli olarak ayarlanmıştır. Model, doğruluğu ve kaybı takip edilerek farklı mimari versiyonlarla toplamda altı deneme sonucu test edilmiştir.

En başarılı sonuç, derinliği artırılmış bir mimari ile elde edilmiş ve test doğruluğu %96.33 seviyesine ulaşmıştır. Bu başarı, modelin hem sahte hem de gerçek görselleri yüksek doğrulukla ayırt edebildiğini göstermektedir.

| Model No     | Test Doğruluğu | Gözlem                                             |
| ------------ | -------------- | -------------------------------------------------- |
| Deneme 1     | %93.72         | İlk yapı hızlı overfitting gösterdi.               |
| Deneme 2     | %95.86         | Düzenlileştirici teknikler uygulandı.              |
| Deneme 3     | %95.80         | Eğitim süresi uzatıldı.                            |
| Deneme 4     | %96.05         | Parametre iyileştirmeleri yapıldı.                 |
| Deneme 5     | %95.71         | Kararlı sonuçlar alındı.                           |
| **Deneme 6** | **%96.33**     | Katman sayısı artırıldı, en iyi sonuç elde edildi. |


Bu modellerden sonuncusu, mobil uygulamada entegre edilecek şekilde seçilmiş ve üretim ortamına uygun hale getirilmiştir.
